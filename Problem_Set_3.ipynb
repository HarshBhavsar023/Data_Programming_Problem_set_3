Occupations
Step 1. Import the necessary libraries
Step 2. Import the dataset from this address.
Step 3. Assign it to a variable called users
Step 4. Discover what is the mean age per occupation
Step 5. Discover the Male ratio per occupation and sort it from the most to the least
Step 6. For each occupation, calculate the minimum and maximum ages
Step 7. For each combination of occupation and sex, calculate the mean age
Step 8. For each occupation present the percentage of women and men

```python
# Step 1: Import the necessary libraries
import pandas as pd

# Step 2: Import the dataset
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
users = pd.read_csv(url, sep='|')

# Step 3: The dataset is now stored in the variable 'users'

# Step 4: Calculate the mean age per occupation
mean_age_by_occupation = users.groupby('occupation')['age'].mean()

# Step 5: Calculate the Male ratio per occupation and sort it
def male_ratio(group):
    return (group['gender'] == 'M').sum() / group['gender'].count()

male_ratio_by_occupation = users.groupby('occupation').apply(male_ratio)
sorted_male_ratio = male_ratio_by_occupation.sort_values(ascending=False)

# Step 6: Calculate the minimum and maximum ages per occupation
age_range_by_occupation = users.groupby('occupation')['age'].agg([min, max])

# Step 7: Calculate the mean age per combination of occupation and sex
mean_age_by_occupation_sex = users.groupby(['occupation', 'gender'])['age'].mean()

# Step 8: Calculate the percentage of women and men per occupation
def gender_percentage(group):
    total_count = group['gender'].count()
    return {
        'percentage_women': (group['gender'] == 'F').sum() / total_count * 100,
        'percentage_men': (group['gender'] == 'M').sum() / total_count * 100
    }

gender_percentage_by_occupation = users.groupby('occupation').apply(gender_percentage)

# Printing the results
print("Mean age per occupation:\n", mean_age_by_occupation)
print("\nMale ratio per occupation (sorted):\n", sorted_male_ratio)
print("\nMinimum and maximum ages per occupation:\n", age_range_by_occupation)
print("\nMean age per combination of occupation and sex:\n", mean_age_by_occupation_sex)
print("\nPercentage of women and men per occupation:\n", gender_percentage_by_occupation)
```

Question 4 : Wind Statistics

```python
# Step 1: Import the necessary libraries
import pandas as pd
import numpy as np

# Step 2: Import the dataset
url = "wind.txt"  # Replace with the actual path to the dataset
data = pd.read_csv(url, delim_whitespace=True, parse_dates=[[0, 1, 2]])

# Step 3: Assign datetime index
data.set_index('Yr_Mo_Dy', inplace=True)

# Step 4: Function to fix year
def fix_year(year):
    if year >= 2000:
        return year - 100
    else:
        return year

data.index = data.index.map(lambda x: x.replace(year=fix_year(x.year)))

# Step 5: Convert index to datetime64[ns]
data.index = pd.to_datetime(data.index)

# Step 6: Count missing values per location
missing_values_per_location = data.isnull().sum()

# Step 7: Count non-missing values
non_missing_values_count = data.count().sum()

# Step 8: Calculate the mean windspeed over all locations and times
mean_windspeed = data.mean().mean()

# Step 9: Calculate location statistics
loc_stats = pd.DataFrame({
    'min': data.min(),
    'max': data.max(),
    'mean': data.mean(),
    'std': data.std()
})

# Step 10: Calculate day statistics
day_stats = pd.DataFrame({
    'min': data.min(axis=1),
    'max': data.max(axis=1),
    'mean': data.mean(axis=1),
    'std': data.std(axis=1)
})

# Step 11: Average windspeed in January for each location
january_avg = data[data.index.month == 1].mean()

# Step 12: Downsample to yearly frequency
yearly_data = data.resample('Y').mean()

# Step 13: Downsample to monthly frequency
monthly_data = data.resample('M').mean()

# Step 14: Downsample to weekly frequency
weekly_data = data.resample('W').mean()

# Step 15: Calculate weekly statistics for the first 52 weeks
weekly_stats = weekly_data.iloc[:52].agg(['min', 'max', 'mean', 'std'])

# Printing the results
print("Missing values per location:\n", missing_values_per_location)
print("\nNumber of non-missing values:", non_missing_values_count)
print("\nMean windspeed over all locations and times:", mean_windspeed)
print("\nLocation statistics:\n", loc_stats)
print("\nDay statistics:\n", day_stats)
print("\nAverage windspeed in January for each location:\n", january_avg)
print("\nYearly data:\n", yearly_data)
print("\nMonthly data:\n", monthly_data)
print("\nWeekly data:\n", weekly_data)
print("\nWeekly statistics for the first 52 weeks:\n", weekly_stats)
```

Question 6 : Create a line plot showing the number of marriages and divorces per capita in the
U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Import the necessary libraries

# Step 2: Load the dataset
file_path = "us-marriages-divorces-1867-2014.csv"
data = pd.read_csv(file_path)

# Step 3: Create a line plot
plt.figure(figsize=(10, 6))

# Plotting the data
plt.plot(data['Year'], data['Marriages'], label='Marriages')
plt.plot(data['Year'], data['Divorces'], label='Divorces')

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Number per Capita')
plt.title('Number of Marriages and Divorces per Capita in the U.S. (1867-2014)')
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

Question 8 : Create a horizontal bar chart that compares the deadliest actors in Hollywood. Sort
the actors by their kill count and label each bar with the corresponding actor's name.
Don't forget to label your axes!

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Import the necessary libraries

# Step 2: Load the dataset
file_path = "actor_kill_counts.csv"
data = pd.read_csv(file_path)

# Step 3: Sort the actors by kill count
data_sorted = data.sort_values(by='Count', ascending=False)

# Step 4: Create a horizontal bar chart
plt.figure(figsize=(10, 6))

# Plotting the data
plt.barh(data_sorted['Actor'], data_sorted['Count'])

# Adding labels and title
plt.xlabel('Count')
plt.ylabel('Actor')
plt.title('Deadliest Actors in Hollywood')
plt.gca().invert_yaxis()  # To have the actor with the highest kill count on top

# Display the plot
plt.grid(True)
plt.show()
```

Question 9 : Create a pie chart showing the fraction of all Roman Emperors that were
assassinated.
Make sure that the pie chart is an even circle, labels the categories, and shows the
percentage breakdown of the categories.

```python
import pandas as pd
emperor = pd.read_csv('roman-emperor-reigns.csv')

a=emperor.loc[emperor.Cause_of_Death.isin(['Assassinated']),['Emperor','Length_of_Reign','Cause_of_Death']]
import matplotlib.pyplot as plt


labels = a.Emperor
y=a.Length_of_Reign.values
font = {"fontsize":24}

fig1, ax1 = plt.subplots(figsize=[25,35])
ax1.pie(y , labels=labels, autopct='%1.1f%%', startangle=90, textprops =font)
ax1.axis('equal')
ax1.set_title("Roman Emperors that were assassinated", fontsize=50, fontweight='bold')
plt.show()
```

Question 10 : Create a scatter plot showing the relationship between the total revenue earned by
arcades and the number of Computer Science PhDs awarded in the U.S. between
2000 and 2009.
Don't forget to label your axes!
Color each dot according to its year.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Import the necessary libraries

# Step 2: Load the dataset
file_path = "arcade-revenue-vs-cs-doctorates.csv"
data = pd.read_csv(file_path)

# Step 3: Create a scatter plot
plt.figure(figsize=(10, 6))

# Plotting the data
plt.scatter(data['Total Arcade Revenue (billions)'], data['Computer Science Doctorates Awarded (US)'], c=data['Year'], cmap='viridis', s=100)

# Adding labels and title
plt.xlabel('Total Arcade Revenue (billions)')
plt.ylabel('Computer Science Doctorates Awarded (US)')
plt.title('Relationship between Arcade Revenue and CS PhDs (2000-2009)')
plt.colorbar(label='Year')

# Display the plot
plt.grid(True)
plt.show()
```
